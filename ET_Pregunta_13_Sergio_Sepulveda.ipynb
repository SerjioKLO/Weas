```python
import numpy as np

A = np.array([[32, 50],
              [10, 40]])

B = np.array([18700, 9500])

solucion = np.linalg.solve(A, B)

x, y = solucion

print(f"El precio de un asiento en clase A es: ${x:.2f}")
print(f"El precio de un asiento en clase B es: ${y:.2f}")
```
